# All Common Regression Machine Learning Algorithms

This document outlines all the commonly used machine learning algorithms for regression problems where we are trying to produce a best-fit line that gives expected value output given inputs.

## Linear Regression

Linear regression multiples an n-dimensional (input dim) weight vector $\vec{w}$ to the input vector (so multiplying each input value by a weight), and adds on an additional bias. This result is the predicted value. 
$$
\hat{y} = \vec{w} \cdot \vec{x} + b
$$

$$
\hat{y} = w_1x_1 + w_2x_2 + \ldots + w_nx_n + b
$$

It uses the mean-squared error (MSE) cost function, which sums up the square of each difference between the true value and predicted value. 
$$
J = MSE = \frac{1}{m} \sum_{i=1}^{m} (\hat{y_i} - y_i)^2
$$

When training, we use: 
$$
\frac{\partial J}{\partial w_k} = \frac{2}{m} \sum_{i=1}^{m} (\hat{y_i} - y_i) \cdot x_k
$$
$$
\frac{\partial J}{\partial b} = \frac{2}{m} \sum_{i=1}^{m} (\hat{y_i} - y_i)
$$


These derivatives are applied to update the weights using gradient descent:
$$
w_{new} = w_{old} - lr \cdot \frac{\partial J}{\partial w}
$$

## SGD

The first variation we might want to make on the linear regression model is not on the equations themselves, but on how many cases we sample for the cost function on update. Basic linear regression samples all $m$ training cases for each update, but especially on larger datasets, that can be too costly. An alternative is Stochastic Gradient Descent (SGD), which samples only 1 example at a time. There is also mini-batch GD, which samples a small batch.

With SGD, m in the cost function is effectively 1, so we can make the following simplifications:

$$
J = (\hat{y_i} - y_i)^2
$$

$$
\frac{\partial J}{\partial w_k} = 2 x_k (\hat{y_i} - y_i) 
$$

$$
\frac{\partial J}{\partial b} = 2 (\hat{y_i} - y_i)
$$

With mini-batch, just replace every occurence of $m$ in the standard linear regression equations with the batch size.

## Regularization

A problem with the standard linear regression equation is the potential for overfitting, when the model makes a regression or prediction line that is way to specific to the training data and doesn't learn to give accurate answers on testing data. One key hallmark of overfitting is large coefficients, as it means the model is making factors highly senstivie in order to capture minor changes that appear in the training data. A standard technique to prevent this is __regularization__, which effectively adds a penalty based on the size of the coefficients.

The most common regularization terms are called L1 and L2. They are defined as follows:
$$
L1 = \alpha \sum_j |w_j|
$$
$$
L2 = \alpha \sum_j w_j^2
$$

Note that $\alpha$ is simply a constant that controls how aggresively we want to regularize (how much we want to prevent overfitting). This term is added to the MSE cost function from regular linear regression.

One thing that is useful to notice is the derivatives of these terms with respect to a weight. The derivative of $|w_j|$ is +1 when $w_j > 0$ and -1 when $w_j < 1$. Thus:

$$
\frac{\partial L1}{\partial w_k} = \pm \alpha 
$$

Meanwhile, the derivative of $w_j^2$ is $2w_j$, so:

$$
\frac{\partial L2}{\partial w_k} = 2\alpha w_k
$$

The L1 regularization is always making the same update no matter the size of $w_j$, while the derivative/updates for the L2 regularization becomes smaller proportional to $w_j$. That's why L1 regularization often leads to many coefficients converging on 0 while L2 leaves more features considered.

Just by knowing the L1 and L2 regularization terms, we have unlocked 3 more algorithms: Lasso, Ridge, and Elastic!
- Lasso uses L1 and modifies the cost function so $J = MSE + L1$
- Ridge uses L2 and modifies the cost function so $J = MSE + L2$
- Elastic uses both and modifies the cost function so $J = MSE + L1 + L2$

There are certain cases where one is better than the other. For example, use Lasso or Elastic when only a few features are important, and use Ridge when all features should have some consideration.
